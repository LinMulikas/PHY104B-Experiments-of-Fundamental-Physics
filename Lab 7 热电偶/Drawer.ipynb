```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
plt.rcParams['font.size'] = '14'
```

```python
Ts = np.arange(30, 75, 5)
U1s = [1.17, 1.37, 1.57, 1.76, 1.97, 2.17, 2.37, 2.57, 2.78]
U2s = [1.17, 1.37, 1.57, 1.78, 1.99, 2.18, 2.40, 2.61, 2.82]
U_30_70 = 1.63
U_0_30 = 1.17
Ts
```

```python
plt.figure(figsize=(16, 10))
plt.scatter(Ts, U1s)
plt.show()
```

```python
from sklearn.linear_model import  LinearRegression
linear_model = LinearRegression()

linear_model.fit(
    np.array(Ts).reshape((-1, 1)), 
    np.array(U1s).reshape((-1, )))

k1 = linear_model.coef_[0]
b1 = linear_model.intercept_
k1, b1
```

```python
U1s_pred = linear_model.predict(np.array(Ts).reshape((-1, 1)))
plt.figure(figsize=(16, 10))
plt.title("U-T 曲线：测量热电偶系数")
plt.xlabel("T/\u00b0C")
plt.ylabel("U/mV")
plt.scatter(Ts, U1s)
plt.plot(Ts, U1s_pred)
plt.legend(["实验数据", "拟合结果：y = {:.4f}*x - {:.4f}".format(k1, abs(b1))])
plt.show()
```
