```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
plt.rcParams['font.size'] = '14'
```

```python
x1 = np.arange(0.5, 4.0, 0.5)
y1 = np.array([0.49, 0.98, 1.5, 1.95, 2.45, 3.0, 3.5])
x1, y1
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_linear = LinearRegression()
model_linear.fit(x1.reshape((-1, 1)), y1.reshape((-1, )))
```

```python
k1 = model_linear.coef_[0]
b1 = model_linear.intercept_
k1, b1
```

```python
x1_line = np.linspace(0.5, 3.5)
y1_pred = k1 * x1_line + b1
```

```python
plt.figure(figsize=(16, 10))
plt.title("信号发生器幅度-示波器测量幅度图")
plt.plot(x1, y1, marker='o')
plt.plot(x1_line, y1_pred)

plt.legend(["真实数据", "拟合结果：y = 1.001*x - 0.0214"])

plt.show()
```
