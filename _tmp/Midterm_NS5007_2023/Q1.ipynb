```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_raw = pd.read_csv('SpinalCord_clustering_rawCounts.csv')
df_trans = pd.read_csv('SpinalCord_clustering_CLR_Transform.csv')
```

```python
arr_raw = df_raw.iloc[:, 1:].to_numpy()
arr_trans = df_trans.iloc[:, 1:].to_numpy()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score
```

```python
scores_kmeans_cali = []
scores_kmeans_cali_trans = []

for i in range(2, 30):
    model_1 = KMeans(n_clusters=i, n_init='auto')
    model_1.fit(arr_raw)
    model_2 = KMeans(n_clusters=i, n_init='auto')
    model_2.fit(arr_trans)
    
    scores_kmeans_cali.append(calinski_harabasz_score(arr_raw, model_1.labels_))
    scores_kmeans_cali_trans.append(calinski_harabasz_score(arr_trans, model_2.labels_))
    
    size_label_1 = []
    size_label_2 = []
    for j in range(i):
        size_label_1.append(np.argwhere(model_1.labels_ == j).shape[0])
        size_label_2.append(np.argwhere(model_2.labels_ == j).shape[0])

    print("K = ", i)
    print(size_label_1)
    print(size_label_2)
    
    
scores_kmeans_cali, scores_kmeans_cali_trans
```

```python
scores_kmeans_davi = []
scores_kmeans_davi_trans = []

for i in range(2, 30):
    model_1 = KMeans(n_clusters=i, n_init='auto')
    model_1.fit(arr_raw)
    model_2 = KMeans(n_clusters=i, n_init='auto')
    model_2.fit(arr_trans)
    scores_kmeans_davi.append(davies_bouldin_score(arr_raw, model_1.labels_))
    scores_kmeans_davi_trans.append(davies_bouldin_score(arr_trans, model_2.labels_))
    
    size_label_1 = []
    size_label_2 = []
    for j in range(i):
        size_label_1.append(np.argwhere(model_1.labels_ == j).shape[0])
        size_label_2.append(np.argwhere(model_2.labels_ == j).shape[0])

    print("K = ", i)
    print(size_label_1)
    print(size_label_2)
    
scores_kmeans_davi, scores_kmeans_davi_trans
```

```python
plt.figure(figsize=(16, 10))
plt.title("The comparasion of cali score")
plt.plot(scores_kmeans_cali, c='blue')
plt.plot(scores_kmeans_cali_trans, c='orange')
plt.legend(['cali score of raw', 'cali score of trans'])
plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.title("The comparasion of davi score")
plt.plot(scores_kmeans_davi, c='blue')
plt.plot(scores_kmeans_davi_trans, c='orange')
plt.legend(['davi score of raw', 'davi score of trans'])
plt.show()
```
